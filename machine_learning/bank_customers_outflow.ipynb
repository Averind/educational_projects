# Исследование причин оттока клиентов банка

Из «Бета-Банка» стали уходить клиенты. Каждый месяц. Немного, но заметно. Банковские маркетологи посчитали: сохранять текущих клиентов дешевле, чем привлекать новых. Требуется спрогнозировать, уйдёт клиент из банка в ближайшее время или нет. Для работы предоставлены исторические данные о поведении клиентов и расторжении договоров с банком.

Необходимо построить модель со значением *F1*-меры не меньшим, чем 0.59.

### Разбивка данных на выборки

Разобьем данные на обучающую, валидационную и тестовую выборки по схеме 3-1-1.

```python
# выделим обучающую выборку из данных

features_train, features_valid_test, target_train, target_valid_test = train_test_split(
    features, target, test_size=0.4, stratify=target, random_state=12345)
```

```python
# разобьем данные на обучающую и тестову

features_valid, features_test, target_valid, target_test = train_test_split(
    features_valid_test, target_valid_test, test_size=0.5, stratify=target_valid_test, random_state=12345)
```

```python
# выведем общую информацию о каждом датафрейме - с обучающей, валидационной и тестовой выборками

dataframes = [features_train, features_valid, features_test]

for dataframe in dataframes:
    dataframe.info()
    print('''
    
    ''')
```

```python
# выведем первые 10 строк каждого датафрейма - с обучающей, валидационной и тестовой выборками

dataframes = [features_train, features_valid, features_test, target_train, target_valid, target_test]

for dataframe in dataframes:
    display(dataframe.head())
```

С выборками все в порядке.

### Создание моделей

Т.к. наша задача относится к задачам классификации, то создадим модели дерева решений, случайного леса и логистической регрессии с перебором гиперпараметров.

#### Дерево решений

```python
# зададим переменные, которым в дальнейшем присвоим лучшую модель и ее параметры 

best_tree_model = None
best_tree_f1 = 0
best_tree_auc = 0
best_tree_depth = 0
```

```python
# создадим пустой список, которому передадим значения F1-меры для дерева каждой глубины для валидационной выборки

tree_f1_valid = []
tree_auc_valid = []
```

Напишем цикл, который выберет лучшую модель (с наибольшим значением F1-меры) в зависимости от глубины решающего дерева.

```python
%%time

for depth in tqdm(range(1, 31)):
    model = DecisionTreeClassifier(max_depth=depth, random_state=12345)
    model.fit(features_train, target_train)
    predictions_valid = model.predict(features_valid)
    
    f1 = f1_score(target_valid, predictions_valid)
    tree_f1_valid.append(f1)
        
    probabilities_valid = model.predict_proba(features_valid)[:, 1]
    auc = roc_auc_score(target_valid, probabilities_valid)
    tree_auc_valid.append(auc)
    
    if f1 > best_tree_f1:
        best_tree_model = model
        best_tree_f1 = f1
        best_tree_auc = auc
        best_tree_depth = depth
```

Построим линейный график зависимости F1-меры модели от глубины дерева решений.

```python
# преобразуем списки с метриками в датафреймы

tree_f1_valid = pd.DataFrame(tree_f1_valid)
tree_auc_valid = pd.DataFrame(tree_auc_valid)

# зададим для него новую индексацию - в соответствии с глубинами деревьев

tree_f1_valid.index = list(range(1, 31))
tree_auc_valid.index = list(range(1, 31))
```

```python
plt.title('зависимость F1-меры модели \n от глубины дерева решений')

plt.plot(tree_f1_valid);

plt.xlim(1)

plt.legend(['валидационная выборка'])

plt.xlabel('глубина дерева', fontsize=10)
plt.ylabel('F1-мера', fontsize=10)

plt.grid()
```

```python
plt.title('зависимость AUC-ROC модели \n от глубины дерева решений')

plt.plot(tree_auc_valid);

plt.xlim(1)

plt.legend(['валидационная выборка'])

plt.xlabel('глубина дерева', fontsize=10)
plt.ylabel('AUC-ROC', fontsize=10)

plt.grid()
```

```python
print('Значение показателя F1 лучшей модели:', best_tree_f1)
print('Значение показателя AUC-ROC лучшей модели:', best_tree_auc)
print('Глубина дерева лучшей модели:', best_tree_depth)
```

Изучим наиболее удачную модель дерева решений - построим график.

```python
# зададим размер графика

fig = plt.figure(figsize=(35,40))

# построим график дерева решений

_ = plot_tree(best_tree_model, 
               feature_names=features.columns, 
               class_names=['Ушел','Остался'],
               filled=True,
               fontsize=25)
```

График дерева решений, ввиду большой глубины дерева, оказался достаточно загроможденным и плохо читаемым. Тем не менее он дает представление о том, как именно работала модель.

Мы проверили модели с глубиной дерева от 1 до 30.

**Наилучшей моделью дерева решений по F1-мере оказалась оная с глубиной дерева, равной 7. Результат составляет 0.58. Значение показателя AUC-ROC лучшей модели: 0.82**.

Скорее всего, результаты для моделей с глубиной дерева меньшей 7 оказались ниже, т.к. имело место недообучение. В свою очередь более низкие результаты для моделей с глубиной дерева большей 7 связаны с переобучением.

#### Случайный лес

```python
# зададим переменные, которым в дальнейшем присвоим лучшую модель и ее параметры 

best_forest_model = None
best_forest_f1 = 0
best_forest_auc = 0
best_forest_est = 0
best_forest_depth = 0
```

## Борьба с дисбалансом

Как было отмечено выше, в данных целевого показателя наблюдается дисбаланс: 0 встречаются в 3.9 раза чаще, чем единицы.

Попробуем решить данную проблему двумя способами: путем уменьшения выборки и путем ее увеличения.

### Уменьшение выборки

Напишем функцию, которая уменьшит количество значений 0 в датафрейме со всеми показателями и сириэс с целевым показателем в 4 раза, чтобы соотношение 0 и 1 стало похоже на 1:1.

```python
def downsample(features, target, fraction):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]
    
    features_downsampled = pd.concat([features_zeros.sample(frac=fraction, random_state=12345)] + [features_ones])
    
    target_downsampled = pd.concat([target_zeros.sample(frac=fraction, random_state=12345)] + [target_ones])
    
    features_downsampled, target_downsampled = shuffle(features_downsampled, target_downsampled, random_state=12345)
    
    return features_downsampled, target_downsampled
```

```python
# запишем переменные с признаками и целевым признаком после уменьшения выборки

features_downsampled, target_downsampled = downsample(
    features_train,
    target_train,
    0.25
)
```

```python
print('Значение показателя F1 лучшей модели:', best_logreg_upsampled_f1)
print('Значение показателя AUC-ROC лучшей модели:', best_logreg_upsampled_auc)
print('Количество итераций обучения лучшей модели:', best_logreg_upsampled_iter)
```

**Наилучшей моделью логистической регрессии по показателю F1-меры оказалась оная с количеством итераций обучения, равной 100. Результат составляет 0.45. AUC-ROC лучшей модели: 0.72**.

До проведения мероприятий по решению проблемы дисбаланса классов наилучшей моделью логистической регрессии оказалась оная с количеством итераций обучения, равной 100. Результат составляет 0.08. AUC-ROC лучшей модели: 0.68.

Таким образом, **уменьшение выборки значительно улучшило результаты**.
